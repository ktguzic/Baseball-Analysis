```python
bb = bb.drop(columns = ['SB', 'CS', 'SO', 'OPS+', 'GDP', 'HBP', 'SH', 'SF', 'IBB'])
```

```python
bb['Pos'].fillna('No field position played', inplace=True)
```

```python
bb['Awards'].fillna('Did not receive', inplace=True)
```
